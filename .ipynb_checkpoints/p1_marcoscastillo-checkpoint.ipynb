```python
# [Título del trabajo/tarea/(módulo del proyecto)] 


| Código | Description |
| ------:| ----------- |
| ***Robotica*** | 01 | 
| **UNAM FI-2025-I** | Tarea *01* |
| **Robotica-2025-I**  | Practica *01* |
| **318094579** | URDF |

Previo 

 1. Definición de URDF
El Universal Robot Description File (URDF) es un formato de archivo XML utilizado en robótica para describir la estructura y propiedades físicas de un robot. URDF permite definir los componentes del robot, como eslabones, juntas y propiedades físicas (masa, inercia, etc.), así como la representación visual y colisiones. Este formato facilita la simulación y el control de robots en plataformas como ROS (Robot Operating System) y es esencial para generar modelos virtuales que se puedan visualizar en herramientas como RViz y simular en plataformas como Gazebo.

2. Tipos de juntas que están definidas dentro del formato URDF
En el formato URDF, se pueden definir varios tipos de juntas que determinan el movimiento relativo entre los eslabones de un robot. Los tipos de juntas más comunes son:

- Revolute (revoluta): Permite el movimiento de rotación alrededor de un eje fijo. Se usa comúnmente para articulaciones que permiten rotación, como las caderas o los codos en robots.

- Prismatic (prismática): Permite el movimiento lineal a lo largo de un eje fijo. Se utiliza para movimientos lineales, como los actuadores de un robot que permiten deslizamientos.

- Fixed (fija): No permite ningún movimiento relativo entre los eslabones. Se utiliza para crear un vínculo rígido entre dos eslabones.

- Continuous (continua): Similar a la junta revolute, pero permite una rotación continua sin un límite angular (por ejemplo, un motor que da vueltas completas sin límite).

3. Definición de elementos primitivos que se pueden utilizar para definir un robot en ROS
En el contexto de ROS (Robot Operating System), los elementos primitivos son las construcciones básicas utilizadas para definir la geometría y la estructura de un robot. Algunos de estos elementos incluyen:

- Link (eslabón): Representa una parte física del robot, como un brazo o una pierna. Se pueden agregar propiedades como la geometría (formas simples como cajas, cilindros, esferas), la visualización, la colisión y la inercia.

- Joint (junta): Define cómo se conectan dos eslabones, permitiendo especificar si el movimiento es de tipo rotacional (revolute), traslacional (prismatic), o rígido (fixed). Además, se pueden incluir limitaciones sobre el movimiento (rangos de movimiento, velocidad, esfuerzo).

- Transmission (transmisión): Define cómo se conecta un actuador (motor, servo, etc.) a una junta, permitiendo controlar el movimiento del robot.

- Sensor: En algunos casos, se incluyen sensores, como cámaras o sensores de distancia, para obtener información sobre el entorno del robot.

4. Qué es un nodo y qué es un tópico en ROS 2
- Nodo: En ROS 2, un nodo es una instancia de un proceso que realiza una tarea específica dentro del sistema de robótica. Los nodos pueden ser responsables de controlar un motor, leer datos de un sensor, procesar imágenes, entre otros. Cada nodo ejecuta una función dentro del ecosistema distribuido de ROS, y se comunican entre sí a través de mensajes y servicios.

- Tópico: Un tópico en ROS 2 es un canal de comunicación mediante el cual los nodos pueden intercambiar mensajes. Los nodos que publican información lo hacen a un tópico, mientras que los nodos que desean recibir esa información se suscriben a dicho tópico. Este mecanismo permite la comunicación asincrónica entre nodos de manera eficiente.

5. Qué es RVIZ
RViz es una herramienta de visualización en 3D utilizada en el ecosistema de ROS. Su principal función es permitir a los usuarios visualizar en tiempo real el estado y los datos de un robot o de un sistema robótico. RViz puede mostrar la posición de los eslabones, las trayectorias planificadas, las lecturas de sensores, como imágenes de cámaras o nubes de puntos de LiDAR, y otros datos relevantes para el control y la simulación del robot. Es una herramienta crucial para la depuración y la validación de robots en entornos de simulación antes de implementarlos en hardware real.

Enlaces de consulta en formato IEEE:
ROS Wiki, URDF, [en línea]. Disponible en: https://wiki.ros.org/urdf.
ROS Wiki, Joints, [en línea]. Disponible en: https://wiki.ros.org/urdf/XML/joint.
ROS Wiki, Primitives, [en línea]. Disponible en: https://wiki.ros.org/urdf/XML/link.
ROS Wiki, ROS 2, [en línea]. Disponible en: https://index.ros.org/doc/ros2/.
ROS Wiki, RViz, [en línea]. Disponible en: https://wiki.ros.org/rviz.
 
 
## Contenido

- [Objetivo](#objetivo)
- [Introducción](#introduccion)
- [Desarrollo](#desarrollo)
- [Conclusiones](#conclusiones)
- [Autor](#autor)
- [Referencias](#referencias)

## Objetivo

El objetivo de esta práctica es que el alumno comprenda, interprete y modifique
la información básica de los formatos de Universal Robot Description File (URDF).


## Introducción

El Universal Robot Description File (URDF) es un formato de archivo ampliamente
utilizado en robótica para describir de manera precisa la estructura y propiedades
de un robot. Este formato permite la definición de eslabones, juntas, propiedades
visuales y físicas, lo que facilita la simulación y el control de robots en plataformas
como ROS (Robot Operating System) [1]. A través de un URDF, es posible crear modelos
virtuales de robots que pueden ser utilizados en simuladores como Gazebo o herramientas
de visualización como RViz. En esencia, un archivo URDF actúa como el plano digital
de un robot, permitiendo su interpretación tanto para simulaciones como para
despliegues reales en hardware. La versatilidad del URDF lo convierte en una pieza
fundamental para ingenieros y desarrolladores en el campo de la robótica.

    
## Desarrollo

En esta practica se define un modelo simple de robot en el formato Universal Robot Description File (URDF). El robot en cuestión consta de tres eslabones (uno fijo y dos móviles) y está
    compuesto por dos grados de libertad, es decir, dos juntas revolutas (o rotacionales). 
    
<robot name="basic_link">

    <!-- Definición del "mundo", que es el marco de referencia principal para el robot -->
    <link name="world">
    </link>

    <!-- Creación de una unión fija entre el "mundo" y el eslabón base del robot -->
    <joint name="world_to_base_link" type="fixed">
        <!-- El eslabón "world" será el padre de esta unión -->
        <parent link="world"/>
        <!-- El eslabón "base_link" será el hijo, conectado al eslabón padre -->
        <child link="base_link"/>
        <!-- Posición y orientación de la junta en relación con los ejes X, Y y Z -->
        <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
    </joint>
    
    <!-- Definición del eslabón base del robot, el cual estará conectado al mundo -->
    <link name="base_link">
        <!-- Información visual del eslabón base, es decir, cómo se verá en la simulación -->
        <visual>
            <!-- Posición inicial y orientación del eslabón visual respecto a su propio sistema de coordenadas -->
            <origin xyz="0.0 0.0 0.075" rpy="0.0 0.0 0.0"/>
            <!-- Descripción de la geometría del eslabón para la representación visual -->
            <geometry>
                <box size="0.1 0.12 0.15"/>             
            </geometry>
            <!-- Definición del material del eslabón, en este caso un color gris -->
            <material name="grey">
                <color rgba="0.5 0.5 0.5 1.0" />
            </material>      
        </visual>
        <!-- Sección que describe las propiedades de colisión del eslabón base -->
        <collision>
            <!-- Posición de la geometría de colisión respecto al sistema de coordenadas del eslabón -->
            <origin xyz="0.0 0.0 0.075" rpy="0.0 0.0 0.0"/>
            <!-- Geometría utilizada para los cálculos de colisión; coincide con la geometría visual -->
            <geometry>
                <box size="0.1 0.12 0.15"/>             
            </geometry>          
        </collision>
        <!-- Propiedades inerciales del eslabón, importantes para la dinámica del sistema -->
        <inertial>
            <!-- Masa del eslabón base -->
            <mass value="0.5"/>
            <!-- Posición del centro de masa respecto al eslabón -->
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
            <!-- Matriz de inercia en torno al centro de masa del eslabón, usada para cálculos de movimiento -->
            <inertia ixx="0.1" ixy="0.0" ixz="0.0" iyy="1.0" iyz="0.0" izz="1.0"/> 
        </inertial>
        
    </link>

    <!-- Definición del primer eslabón móvil del robot -->
    <link name="link_1">
        <!-- Aspecto visual del eslabón, que se verá en el simulador -->
        <visual>
            <!-- Establece la posición y orientación del eslabón visual en relación con su origen -->
            <origin xyz="0.025 0.0 0.25" rpy="0.0 0.0 0.0"/>
            <!-- Geometría del eslabón, que en este caso es una caja -->
            <geometry>
                <box size="0.05 0.05 0.5"/>             
            </geometry>
            <!-- Material usado para el eslabón; en este caso, es un color azul claro -->
            <material name="light_blue">
                <color rgba="0.0 0.75 1.0 1.0" />
            </material>      
        </visual>
        <!-- Define las propiedades de colisión de este eslabón -->
        <collision>
            <!-- Posición de la geometría de colisión respecto al origen del eslabón -->
            <origin xyz="0.025 0.0 0.25" rpy="0.0 0.0 0.0"/>
            <!-- Geometría usada para las interacciones físicas en simulación, como colisiones -->
            <geometry>
                <box size="0.05 0.05 0.5"/>             
            </geometry>          
        </collision>
        <!-- Sección que describe la masa y otras propiedades dinámicas del eslabón -->
        <inertial>
            <!-- Masa asignada al eslabón móvil -->
            <mass value="0.3"/>
            <!-- Ubicación del sistema inercial del eslabón, que normalmente coincide con su centro de masa -->
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
            <!-- La matriz de inercia en el centro de masa, usada para el cálculo de dinámica -->
            <inertia ixx="0.1" ixy="0.0" ixz="0.0" iyy="1.0" iyz="0.0" izz="1.0"/> 
        </inertial>
    </link>

    <!-- Definición de la junta que conecta el eslabón base con el primer eslabón móvil -->
    <joint name="base_link_to_link_1" type="revolute">
        <!-- Define el eslabón padre (base) de esta junta -->
        <parent link="base_link"/>
        <!-- Define el eslabón hijo (primer eslabón móvil) conectado a la base -->
        <child link="link_1"/>
        <!-- Eje sobre el cual se permite el movimiento de rotación de la junta -->
        <axis xyz="1 0 0"/>
        <!-- Límites de la junta en términos de posición angular, velocidad y esfuerzo máximo permitido -->
        <limit lower="-3.14159" upper="3.14159" velocity="50.0" effort="1000.0"/>
        <!-- Posición de la junta en el espacio del eslabón base -->
        <origin xyz="0.05 0 0.15" rpy="0 0 0"/>
    </joint>

    <!-- Definición del segundo eslabón del robot -->
    <link name="link_2">
        <!-- Parte visual que describe cómo se verá el segundo eslabón en el simulador -->
        <visual>
            <!-- Posición y orientación del eslabón visual respecto a su origen -->
            <origin xyz="0.030 0.00 0.25" rpy="0.0 0.0 0.0"/>
            <!-- Geometría usada para la representación gráfica del eslabón, en este caso una caja -->
            <geometry>
                <box size="0.05 0.05 0.5"/>             
            </geometry>
            <!-- El color naranja que representará este eslabón en el simulador -->
            <material name="orange">
                <color rgba="1.0 0.5 0.0 1.0" />
            </material>      
        </visual>
        <!-- Definición de las propiedades de colisión del eslabón, importante para la simulación física -->
        <collision>
            <!-- Establece la posición de la geometría de colisión del eslabón -->
            <origin xyz="0.030 0.05 0.25" rpy="0.0 0.0 0.0"/>
            <!-- Geometría utilizada para detectar colisiones durante la simulación -->
            <geometry>
                <box size="0.05 0.05 0.5"/>             
            </geometry>          
        </collision>
        <!-- Propiedades inerciales necesarias para los cálculos dinámicos en simulación -->
        <inertial>
            <!-- Masa del segundo eslabón -->
            <mass value="0.3"/>
            <!-- Posición del centro de masa del eslabón -->
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
            <!-- Matriz de inercia, que se usa para calcular cómo responde el eslabón al movimiento -->
            <inertia ixx="0.1" ixy="0.0" ixz="0.0" iyy="1.0" iyz="0.0" izz="1.0"/> 
        </inertial>
    </link>

    <!-- Junta revoluta que conecta el primer eslabón móvil con el segundo -->
    <joint name="link_1_to_link_2" type="revolute">
        <!-- Define el primer eslabón como el padre en esta conexión -->
        <parent link="link_1"/>
        <!-- El segundo eslabón será el hijo conectado a la junta -->
        <child link="link_2"/>
        <!-- Eje de rotación permitido para la junta (en este caso, alrededor del eje X) -->
        <axis xyz="1 0 0"/>
        <!-- Límites de movimiento angular, velocidad y esfuerzo máximo que puede soportar la junta -->
        <limit lower="-3.14159" upper="3.14159" velocity="50.0" effort="1000.0"/>
        <!-- Posición de la junta respecto al primer eslabón móvil -->
        <origin xyz="0.05 0 0.5" rpy="0 0 0"/>
    </joint>

    <!-- Definición de las propiedades en Gazebo para el simulador -->
    <gazebo reference="base_link">
        <material>Gazebo/Grey</material>
    </gazebo>

    <gazebo reference="link_1">
        <material>Gazebo/LightBlue</material>
    </gazebo>

    <!-- Fin de las propiedades relacionadas con Gazebo -->

    <!-- Configuración de los elementos de control del robot -->
    <transmission name ="trans_1">
        <!-- Transmisión simple que conecta un motor a una junta -->
        <type>transmission_interface/SimpleTransmission</type>
        <joint name="base_link_to_link_1">
            <!-- Interfaz de hardware para el control de la posición de la junta -->
            <hardwareInterface>hardware_interface/PositionJointInterface</hardwareInterface>
        </joint>
        <!-- Actuador que impulsa la junta -->
        <actuator name="motor_1">
            <hardwareInterface>hardware_interface/PositionJointInterface</hardwareInterface>
            <mechanicalReduction>1</mechanicalReduction>
        </actuator>
    </transmission>

    <!-- Plugin que publica el estado de la junta en el simulador Gazebo -->
    <gazebo>
        <plugin name="joint_state_publisher_controller"
                filename="libgazebo_ros_joint_state_publisher.so">
            <!-- Tasa de actualización del estado de la junta (en Hz) -->
            <update_rate>10</update_rate>
            <!-- Nombre de la junta cuyo estado se publicará -->
            <joint_name>base_link_to_link_1</joint_name>
        </plugin>
    </gazebo>

    <!-- Plugin para el control de trayectoria de la junta en Gazebo -->
    <gazebo>
        <plugin name="joint_pose_trajectory_controller"
                filename="libgazebo_ros_joint_pose_trajectory.so">
            <!-- Frecuencia de actualización del controlador -->
            <update_rate>2</update_rate>
        </plugin>
    </gazebo>

    <!-- Fin de la sección de control del robot -->

</robot>
10:03 p. m.


Bloques de cita

1. ROS Wiki, "URDF," [en línea]. Disponible en: https://wiki.ros.org/urdf.

2. Gazebo Simulator, "Gazebo," [en línea]. Disponible en: https://gazebosim.org/.

3. ROS Wiki, "ros_control," [en línea]. Disponible en: https://wiki.ros.org/ros_control.

4. ROS Wiki, "RViz," [en línea]. Disponible en: https://wiki.ros.org/rviz.

                      
## Conclusiones

El URDF es una herramienta fundamental para los ingenieros y desarrolladores en robótica. Este formato facilita la creación de modelos virtuales de robots que pueden ser utilizados en simuladores como Gazebo y herramientas de visualización como RViz. A través de la definición de eslabones, juntas y propiedades físicas, los usuarios pueden experimentar con el comportamiento de los robots en entornos simulados antes de implementarlos en hardware real.
Este código URDF en particular describe un robot de dos grados de libertad, adecuado para simular un brazo robótico básico. A medida que los robots se hacen más complejos, el uso de URDF se vuelve esencial para simular sus dinámicas y verificar su comportamiento antes de la construcción física.

## Autor

***Castillo Ramos Marcos Daniel***
```
